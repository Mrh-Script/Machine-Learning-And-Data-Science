## 3. Reading and Writing JSON Data

### Load a JSON String into a DataFrame

```python
#JSON string
Data = Data = '{"employee_name": "James", "email": "james@gmail.com", "job_profile": [{"title1":"Team Lead", "title2":"Sr. Developer"}]}'

# Load the JSON string into a pandas DataFrame
df = pd.read_json(StringIO(Data))

df
```

### Convert the DataFrame Back to a JSON String

```python
df.to_json()
```

### Convert to JSON with `index` Orientation

```python
df.to_json(orient='index')
```

### Convert to JSON with `records` Orientation

```python
df.to_json(orient='records')
```

## 4. Reading CSV Data from a URL

### Load the Wine Dataset from the UCI Repository

```python
import pandas as pd

# Load the Wine dataset from the UCI Machine Learning Repository
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data", header=None)
```

### Preview the Data

```python
df.head()
```

### Save the DataFrame to a Local CSV File

```python
df.to_csv("wine.csv")
```

## 5. Reading HTML Tables from a URL

### Read Tables from the FDIC Failed Bank List Page

```python
url = "https://www.fdic.gov/resources/resolutions/bank-failures/failed-bank-list/"

df = pd.read_html(url)
```

### View the First Table Found

```python
df[0]
```

### Read a Specific Table by Matching Text (Wikipedia: Mobile Country Codes)

```python
url = "https://en.wikipedia.org/wiki/Mobile_country_codes"

pd.read_html(url, match="Country",header=0, storage_options={"User-Agent": "Mozilla/5.0"})[0]
```

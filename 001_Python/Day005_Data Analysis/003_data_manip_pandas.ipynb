A walkthrough of common pandas operations: inspecting data, handling missing values, renaming columns, changing data types, grouping/aggregating, and merging DataFrames.

```python
import pandas as pd
```

## 3. Handling Missing Values

```python
df.isnull()

# Missing = True     |     Not Missing = False
```

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

### Filling Missing Values with a Constant

```python
## Fill the missing values with data

df_filled = df.fillna(0)    # replace with 0
df_filled 
```

### Filling Missing Values with the Column Mean

```python
### Filling missing values with the mean of the column

## Fill the missing sales with it's mean
df['Sales_fillNA'] = df['Sales'].fillna(df['Sales'].mean())

df      # Creates new Collumn - Sales_fillNA : and stores mean of Sales
```

```python
df.dtypes
```

## 4. Renaming Columns

```python
df = df.rename(columns = {'Date' : 'Sales Date'})

df.head(3)
```

## 5. Changing Data Types

```python
df['value_new'] = df['Value'].fillna(df['Value'].mean()).astype(int)

df.head()
```

### Creating a New Column from an Existing One

```python
df['New Value'] = df['Value'].apply(lambda x:x*2)

df.head()
```

## 6. Grouping and Aggregating Data

```python
df.head()
```

### Grouping by a Single Column

```python
grouped_mean = df.groupby('Product')['Value'].mean()

print(grouped_mean)
```

### Grouping by Multiple Columns

```python
grouped_sum = df.groupby(['Product', 'Region'])['Value'].sum()

print(grouped_sum)
```

```python
df.groupby(['Product', 'Region'])['Value'].mean()
```

### Aggregating with Multiple Functions

```python
##  Agregate Multiple Functions

grouped_agg = df.groupby('Region')['Value'].agg(['sum', 'mean', 'count'])

grouped_agg
```

## 7. Merging and Joining DataFrames

### Sample DataFrames

```python
# Create a simple DataFrames

df1 = pd.DataFrame({'Key' : ['A', 'B', 'C'], 'Value1': [1,2,3]})
df2 = pd.DataFrame({'Key' : ['A', 'B', 'D'], 'Value2': [4,5,6]})
```

```python
df1
```

```python
df2
```

### Inner Join

```python
## Marge DataFrame on the 'Key Columns'

pd.merge(df1, df2, on="Key", how="inner")    ## Inner Join
```

### Outer Join

```python
pd.merge(df1, df2, on="Key", how="outer")    ## Outer Join
```

### Left Join

```python
pd.merge(df1, df2, on="Key", how="left")     ## Left Outer Join
```

### Right Join

```python
pd.merge(df1, df2, on="Key", how="right")     ## Right Outer Join
```

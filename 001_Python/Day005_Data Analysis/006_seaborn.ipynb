### Import Seaborn
- Import seaborn, conventionally aliased as `sns`

```python
import seaborn as sns
```

### Load Sample Dataset
- Loads the built-in `tips` dataset bundled with seaborn

```python
### Basic ploting with Seaborn

tips = sns.load_dataset('tips')
tips
```

### Scatter Plot
- Visualizes the relationship between `total_bill` and `tip`

```python
### Create a Scatter Plot

import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y = 'tip', data = tips)
plt.title("Sactter Plot of Total Bills vs Tip")
plt.show()
```

### Line Plot
- Shows how `total_bill` trends across party `size`

```python
### Line Plot

sns.lineplot(x='size', y='total_bill', data=tips)
plt.title("Line Plot of Total bill by Size")
plt.show()
```

### Bar Plot – Categorical Comparison
- Compares average `total_bill` across days of the week

```python
### Categorical Plots

sns.barplot(x='day', y='total_bill', data=tips)

plt.title("Bar Plot of Total Bill by Day")
plt.show()
```

### Box Plot
- Shows distribution, median, and outliers of `total_bill` by day

```python
## Box Plot

sns.boxplot(x='day', y='total_bill', data=tips)
```

### Violin Plot
- Combines a box plot and KDE to show the distribution shape of `total_bill` by day

```python
## Violin Plot

sns.violinplot(x='day', y='total_bill', data=tips)
```

### Histogram (no KDE)
- Distribution of `total_bill` split into 10 bins

```python
## Histograms

sns.histplot(tips['total_bill'], bins=10, kde=False)
```

### Histogram with KDE Overlay
- Same histogram with a kernel density estimate curve added

```python
sns.histplot(tips['total_bill'], bins=10, kde=True)
```

### KDE Plot (unfilled)
- Smooth density estimate of `total_bill`

```python
## KDE Plot

sns.kdeplot(tips['total_bill'], fill=False)
```

### KDE Plot (filled)
- Same KDE plot with the area under the curve filled in

```python
sns.kdeplot(tips['total_bill'], fill=True)
```

### Pair Plot
- Grid of pairwise scatter plots and histograms for all numeric columns in `tips`

```python
## Pairplot

sns.pairplot(tips)
```

### Correlation Matrix
- Computes correlation between `total_bill`, `tip`, and `size`

```python
## Heatmap

corr = tips[['total_bill', 'tip', 'size']].corr()

corr
```

### Heatmap of Correlations
- Visualizes the correlation matrix with annotated values

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

### Load Sales Data
- Reads `sales_data.csv` into a pandas DataFrame and previews the first rows

```python
import pandas as pd

sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

### Total Sales by Product Category
- Bar plot summing `Total Revenue` for each product category

```python
## Plot Total Sales by Product

plt.figure(figsize=(10,6))
sns.barplot(x='Product Category', y='Total Revenue', data = sales_df, estimator=sum)

plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")

plt.show()
```

### Total Sales by Region
- Bar plot summing `Total Revenue` for each region

```python
## Plot Total Sales by Region

plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Total Revenue', data = sales_df, estimator=sum)

plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")

plt.show()
```

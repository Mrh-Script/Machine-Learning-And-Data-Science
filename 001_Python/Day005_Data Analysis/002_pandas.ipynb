# Pandas

- **Pandas** is a powerful data manipulation and analysis library in Python.
- It is widely used for:
  - Data analysis
  - Data cleaning
  - Data manipulation

## Primary Data Structures

Pandas provides two main data structures:

### 1. Series

- A **one-dimensional** array-like object.
- Stores a single column of data.
- Each element has an associated index (label).

### 2. DataFrame

- A **two-dimensional**, size-mutable, and potentially heterogeneous tabular data structure.
- Organized into **rows and columns** with labeled axes.
- Can store different data types in different columns.

## Comparison

| Feature | Series | DataFrame |
|---------|--------|-----------|
| Dimension | 1D | 2D |
| Structure | Single column | Rows and columns |
| Data Types | Usually one type | Multiple data types |
| Index | One axis (rows) | Two axes (rows and columns) |

## 01. Series

```python
# A Pandas Series is a one-dimensional array-like object that can hold any data type. It is similar to a column in a table

data = [1, 2, 3, 4, 5]
series = pd.Series(data)

print("Series \n", series)    # left side : default indexes  | right side : values
print(type(series))
```

```python
# Create a Series from dictionary

data = {'a': 1, 'b': 2, 'c': 3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
# Marge data and index 

data = [10,20,30]
index = ['a', 'b', 'c']

print(pd.Series(data, index=index))
```

## 02. DataFrame

```python
# Create a dataframe from a dictionary of list

data = {
    'Name' : ['Riad', 'Sajib', 'Shakib'],
    'Age' : [25, 26, 23],
    'City' : ['Dhaka', 'Rajshahi', 'Joypurhat']
}

df = pd.DataFrame(data)

print(df)
print(type(df))
```

```python
import numpy as np

print(np.array(df))
```

```python
# Create a Data frame From a List of Dictionaries

data = [
    {'Name': 'Riad', 'Age': 25, 'City': 'Dhaka'},
    {'Name': 'Sajib', 'Age': 26, 'City': 'Rajshahi'},
    {'Name': 'Shakib', 'Age': 23, 'City': 'Joypurhat'},
    {'Name': 'Pritom', 'Age': 26, 'City': 'Bogura'}
]

df = pd.DataFrame(data)
print(df)
print(type(df))
```

#### Accessing Data from DataFrame

```python
df
```

```python
# Accessing Attributes

print(df['Name'])

print(type(df['Name']))
```

```python
# Accessing Row

df.loc[1]    # loc[row]  | loc = location
```

```python
# Accessing Specific Cell Data

print(df.iloc[0,0])   # Riad     |   iloc = index location
print(df.iloc[3,2])   # Bogura
```

```python
# Accessing Specific Element
 
print(df.at[2, 'City'] )    # 2nd row of City
print(df.at[3, 'Age'] )    # 3rd row of Age
```

```python
# Another way - using "iat"

df.iat[0,2]   # iat[row, col]
```

## 03. Data Manipulation with DataFrame

```python
df
```

```python
# Adding a Column

df['Salary'] = [40000, 35000, 45000, 35000]

df
```

####  X-Axis = Row = 0    
####  Y-Axis = Column = 1

```python
# Remove a column  (Temporary)

df.drop('Salary',axis=1)     # Salary is a column

# drop is not permanent operation. So not removed permanently
```

```python
# Prove of Temporary Deletion
df
```

```python
# Remove a column Permanently

# inplace = 1

df.drop('Salary', axis=1, inplace = True)

df
```

```python
## Add(+)  with Column

df["Age"] = df['Age'] + 5
df
```

```python
# Remove a Row Permanently

df.drop(2, inplace=True)

df
```

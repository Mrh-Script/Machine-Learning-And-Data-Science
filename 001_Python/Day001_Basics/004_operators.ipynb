## Arithmatic Operators Example

```python
a = 10
b = 5

add_result = a + b         # addition
sub_result = a - b         # subtraction
mult_result = a * b        # multiplication
div_result = a / b         # division
modulus_result = a % b     # modulus
exponent_result = a ** b   # exponentiation

floor_div_result = a // b  # floor division

print("Addition :", add_result)
print("Subtraction :", sub_result)
print("Multiplication :", mult_result)
print("Division :", div_result)
print("Modulus :", modulus_result)
print("Exponentiation :", exponent_result)
print("Floor Division :", floor_div_result)
```

## Comparison Operators Example

```python
a = 10
b = 5
c = 10

# Equal to (==)
print("a == c :", a == c)       # Checks if two values are equal

# Not equal to (!=)
print("a != b :", a != b)       # Checks if two values are not equal

# Greater than (>)
print("a > b :", a > b)         # Checks if left value is greater than right value

# Less than (<)
print("a < b :", a < b)         # Checks if left value is smaller than right value

# Greater than or equal to (>=)
print("a >= c :", a >= c)       # Checks if left value is greater than or equal to right value

# Less than or equal to (<=)
print("b <= a :", b <= a)       # Checks if left value is less than or equal to right value
```

## Logical Operators Example

```python
X = False
Y = True

# AND Operator  : If one input is False, then whole output is False
and_result = X and Y
print("X and Y :", and_result)

# OR Operator  : If one input is True, then whole output is True
or_result = X or Y
print("X or Y :", or_result)

# NOT Operator : Reverses the boolean value
not_result_X = not X
print("not X :", not_result_X)

not_result_Y = not Y
print("not Y :", not_result_Y)
```

## 1. Declaring & Assigning Variables

```python
## Declaring and Assigning

name = "Riad"
age = 25
bmi = 18.7
is_Student = True

## printing the variables

print("Name : ", name)
print("Age : ", age)
print("BMI : ", bmi)
```

## 2. Naming Conventions

- variable name should be descriptive
- They must start with a letter or an underscore '_' and  contains letter, numbers and underscore
- Variable name case-sensitive

```python
## Valid Variable names

age = 25
_name = "Riad"
bmi2 = 18.7
```

```python
## Invalid Variable names

2age = 25
first-name = "Riad"
```

```python
## Case Sensivity
name = "Riad"
Name = "Hasan"

print(name)
print(Name)
```

## 3. Type Casting

```python
age = 25
print(type(age))

float_age = float(age)
print(float_age)

print(type(float_age))
```

## 4. Dynamic Typing

```python
var = 10
print(var, type(var))

var = "Bangladesh"
print(var, type(var))

var = 3.1416
print(var, type(var))
```

## 5. Taking Input
- By default takes values as String
- Need to type cast according to need

```python
## Without Type Casting
age = input("What is your Age?")
print(age, type(age))

## Using Type Casting
intAge = int(input("What is your age?"))
print(intAge, type(intAge))
```

## 6. Basic Calculator Example

```python
value1 = float(input("Enter value 1 : "))
value2 = float(input("Enter value 2 : "))

sum = value1 + value2
difference = value1 - value2
product = value1 * value2
quotient = value1 / value2

print("Sum : ", sum)
print("Difference : ", difference)
print("Product : ", product)
print("Quotient: ", quotient)

```

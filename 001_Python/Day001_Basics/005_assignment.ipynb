**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Using For Loop

num = 5
fact = 1

for i in range(num, 1, -1):
    fact *= i

print("Factorial (For Loop) :", fact)


# Using While Loop

num2 = 6
fact1 = 1
j = num2

while(j>1):
    fact1 *= j
    j -= 1

print("Factorial (While Loop) :", fact1)
```

**Question 16:** Write a Python program to find the sum of the first n natural numbers.

```python
## Way 1 : Using Formula
n = int(input("Enter the value of n : "))
sum = n*(n+1)/2
print("Sum of first ", n ," numbers = ", sum)
```

```python
## Way 2 : Using loop
n = int(input("Enter the value of n : "))
sum = 0
for i in range(1, n+1, +1):
    sum += i

print("Sum of first ", n ," numbers = ", sum)
```

**Question 17:** Write a Python program to check if a year is a leap year.

```python
year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a Leap Year")
else:
    print(year, "is NOT a Leap Year")
```

**Question 19:** Write a Python program to check if a string is a palindrome.

```python
word = input("Word : ")

word = word.replace(" ", "").lower()

if(word == word[::-1]):
    print(word, " is Palindrome!")
else:
    print(word, " is not Palindrome!")
```

**Question 20:** Write a Python program to sort a list of numbers in ascending order.

- sort() : modifies original list
- sorted() : created a new sorted list

```python
numbers = [5, 2, 9, 1, 7, 3]
print("Original list:", numbers)

# Using sort() 
list1 = numbers.copy()   # copy to keep original safe

list1.sort()
print("Ascending order using sort():", list1)

# Using sorted() - does not change original list
list2 = sorted(numbers)
print("Ascending order using sorted():", list2)
```

## 2. Case Sensitive 

```python
name = "Muhammad Riad Hasan"
Name = "From Bangladesh"

print(name)
print(Name)
```

## 3. Indentation
- In statement blocks, use multiple (usually 4) spaces after statements.

```python
age = 32
if age > 30:
    print("Indentation block age : " ,age)

print("Outside indentation block age : " ,age)
```

## 4. Line Continuation


```python
## --> Use (\) to continue the code's statements
total = 1+2+3+4+5\
+6+7

print(total)
```

## 5. Type Inference

```python
## Dynamic data type is possible in Python

variable = 1
print(type(variable))

variable = "Riad"      ## Same variable
print(type(variable))
```

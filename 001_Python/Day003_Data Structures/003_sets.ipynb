# Sets
- Sets are built-in data types in Python used to store collections of unique items. 
- They are unsorted 
- They don't allow duplicate elements

### Uses :
- Used in membership tests
- Eliminating duplicate entries
- Performs mathematical operations : Union, Intersection, Difference, Symmetric difference

```python
# Create a set

my_set = {1, 2, 3, 4, 5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set = set()
print(type(my_empty_set)) 
```

```python
my_set = set([1, 2, 3, 4, 5, 6])
print(my_set)
```

### Don't allow duplicate elements

```python
my_set = {1, 2, 2, 1}
print(my_set)
```

### Add and Remove

```python
my_set = {1,2,4,5,3,6}
print(my_set)
```

```python
## Adding

my_set.add(9)
print(my_set)
```

```python
## Remove

my_set.remove(3)
print(my_set)
```

```python
## If element is not in set

my_set.remove(0)
print(my_set)
```

#### To avoid error, use this method to remove elements from a set ::: "discard" -> Remove the element if it is a member.

```python
print(my_set)
```

```python
my_set.discard(10)
print(my_set)         # No error
```

#### 1. Remove Random element 

```python
print(my_set)

removed_element = my_set.pop()
print(removed_element)

print(my_set)
```

#### 2. Clear a set

```python
print(my_set)

my_set.clear()

print(my_set)
```

#### 3. Set Membership Test

```python
my_set = {1, 2, 3, 4, 5, 6, 7}
print(my_set)

print(4 in my_set)    # True
print(10 in my_set)    # False
```

#### 4. Mathematical Operations

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}
```

##### a) Union

```python
union_set = set1.union(set2)
print(union_set)
```

##### b) Intersection

```python
intersect_set = set1.intersection(set2)
print(intersect_set)
```

##### c) Intersection Update 
- Intersect and then update the first set according to intersection output

```python
intUpdate_set = set1.intersection_update(set2)
print(intUpdate_set)


print(set1)   # set 1 updated
```

##### d) Difference

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}
```

```python
difference_set = set1.difference(set2)
print(difference_set)
```

##### e) Symmetric Difference
- common elements remoeves

```python
symmetricDiff_set = set1.symmetric_difference(set2)

print(symmetricDiff_set)
```

##### f) Is Subset?
- all the elements are common or not

```python
set1 = {1,2,3}
set2 = {1,2,3}

print(set1.issubset(set2))
```

##### g) Super Set
- common either all or less

```python
set1 = {1,2,3}
set2 = {1,2}

print(set1.issuperset(set2))
```

### Q2: Count the unique words in a text.
- Text: We are Bangladeshi. As a Bangladeshi, we speak Bangla.

```python
text = "we are Bangladeshi. As a Bangladeshi, we speak Bangla"

words = text.split()

unique_words = set(words)
print(unique_words)
print("Total Unique words: ", len(unique_words))
```

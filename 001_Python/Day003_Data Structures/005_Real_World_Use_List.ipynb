# Real-World Example using List in Python

### Example 1 : Manage a To-Do List

```python
todo_list = []

while True:
    print("\n===== TO-DO LIST =====")
    print("1. View Tasks")
    print("2. Add Task")
    print("3. Remove Task")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == "1":
        if len(todo_list) == 0:
            print("No tasks available.")
        else:
            print("\nYour Tasks:")
            for i, task in enumerate(todo_list, start=1):
                print(f"{i}. {task}")

    elif choice == "2":
        task = input("Enter a new task: ")
        todo_list.append(task)
        print("Task added successfully!")

    elif choice == "3":
        if len(todo_list) == 0:
            print("No tasks to remove.")
        else:
            print("\nYour Tasks:")
            for i, task in enumerate(todo_list, start=1):
                print(f"{i}. {task}")

            try:
                task_num = int(input("Enter task number to remove: "))
                if 1 <= task_num <= len(todo_list):
                    removed = todo_list.pop(task_num - 1)
                    print(f"Removed: {removed}")
                else:
                    print("Invalid task number.")
            except ValueError:
                print("Please enter a valid number.")

    elif choice == "4":
        print("Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.")
```

### Example 2 : Organizing Student Grades
- A list to store and calculate average grades for students

```python
grades = []

n = int(input("Enter the number of students: "))

for i in range(n):
    grade = float(input(f"Enter grade for student {i + 1}: "))
    grades.append(grade)

# Append new grades
while True:
    choice = input("\nDo you want to add another grade? (yes/no): ").lower()

    if choice == "yes":
        new_grade = float(input("Enter the new grade: "))
        grades.append(new_grade)
        print("Grade added successfully!")

    elif choice == "no":
        break

    else:
        print("Invalid input. Please enter 'yes' or 'no'.")

# Display all grades
print("\nStudent Grades:")
for i, grade in enumerate(grades, start=1):
    print(f"Student {i}: {grade}")

# Calculate statistics
average = sum(grades) / len(grades)
highest = max(grades)
lowest = min(grades)

print(f"\nAverage Grade : {average:.2f}")
print(f"Highest Grade : {highest}")
print(f"Lowest Grade  : {lowest}")
```

### Example 3 : Managing An Inventory for a shop

```python
inventory = ["Rice", "Sugar", "Oil", "Salt", "Tea"]

while True:
    print("\n===== SHOP INVENTORY =====")
    print("1. View Inventory")
    print("2. Add Item")
    print("3. Remove Item")
    print("4. Check Item Availability")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == "1":
        if len(inventory) == 0:
            print("Inventory is empty.")
        else:
            print("\nAvailable Items:")
            for i, item in enumerate(inventory, start=1):
                print(f"{i}. {item}")

    elif choice == "2":
        item = input("Enter item to add: ").title()
        inventory.append(item)
        print(f"{item} added successfully!")

    elif choice == "3":
        item = input("Enter item to remove: ").title()
        if item in inventory:
            inventory.remove(item)
            print(f"{item} removed successfully!")
        else:
            print(f"{item} is not in the inventory.")

    elif choice == "4":
        item = input("Enter item to search: ").title()
        if item in inventory:
            print(f"{item} is available in the inventory.")
        else:
            print(f"{item} is NOT available in the inventory.")

    elif choice == "5":
        print("Thank you!")
        break

    else:
        print("Invalid choice. Please try again.")
```

### Example 4 : Collecting User Feedback

```python
feedback = [
    "Great service!",
    "Very satisfied",
    "Could be better",
    "Excellent experience"
]

# Adding new feedback
feedback.append("Not happy with the service")

# Counting positive feedback
positive_feedback_count = sum(
    1
    for comment in feedback
    if "great" in comment.lower() or "excellent" in comment.lower()
)

print(f"Positive Feedback Count: {positive_feedback_count}")

# Printing all feedback
print("\nUser Feedback:")
for comment in feedback:
    print(f"- {comment}")
```

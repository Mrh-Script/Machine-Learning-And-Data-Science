# Dictionaries
- Unordered collection of items
- Stores data in key-value pairs; Keys must be unique and immutable. (e.g. string, numbers, tuples)
- Values can be of any type

##### 1. Creating Dictionaries

```python
# Way 1
empty_dict = {}
print(type(empty_dict))


# Way 2
empty_dict2 = dict()
print(type(empty_dict2))
```

```python
student = {"name" : "Riad",   # name, age, study are key
           "age" : "25",
           "Study" : "BSc"}

print(student)
print(type(student))
```

```python
# Error (Unique Key)
student = {"name" : "Riad",   
           "age" : "25",
           "Study" : "BSc",
           "name" : "Hasan"}   # name key again

print(student)  # new name key value will reeplace the previous
```

##### 2. Accessing Dictionaries Elements

```python
student = {"name" : "Riad",   
           "age" : "25",
           "Study" : "BSc",
           "grade" : "A+"}

# Way 1 : Regular Way
print(student['name'])
print(student['grade'])

# Way 2 : Using - "get method"
print(student.get('Study'))
print(student.get('address'))    # not in dictionaries

#Define if (not available) - can use any text
print(student.get('contact', "Not available"))  # can use anything
print(student.get('email', "Not exist"))
```

##### 3. Add, Update, Delete

```python
student = {"name" : "Riad",   
           "age" : "25",
           "Study" : "BSc",
           "grade" : "A+"}

print(student)

#Update
student['age'] = "23"
print(student)

# Add
student["address"] = "Bangladesh"
print(student)

# Delete
del student['Study']   # just give key, entire value will delete
print(student)
```

##### 4. Dictionary Method
- finding all the keys
- finding all values
- finding all (key, value) pairs  -> (Items)

```python
student = {"name" : "Riad",   
           "age" : "25",
           "Study" : "BSc",
           "grade" : "A+"}

# Get all the Keys
keys = student.keys()  
print(keys)

# Get all the Values
values = student.values()
print(values)

#Get all pairs (Items)
items = student.items()
print(items)
```

##### 5. Copying Dictionaries (Reference vs Shallow Copy)

```python
# Problem in Copy (this is NOT a copy, just another reference to the same dict)

student_copy = student
print(student)
print(student_copy)
```

```python
# Now Access from both

student["name"] = "Hasan"   #Update

print(student)     # should be update
print(student_copy)   #should not be update

#But both updated because student_copy is not a separate dict,
#it's just another name pointing to the SAME dict object as student
```

```python
# Solution for Copy (Actual Shallow Copy)

student_copy1 = student.copy()  # Shallow Copy, uses a different memory address (new dict object)
print(student_copy1)
print(student)

#Update
student["name"] = "Riad"


print(student)  # Will update
print(student_copy1)  # Will not Update
```

**Note:** `.copy()` only creates a *shallow* copy — it copies the top-level keys/values into a new dict, but if a value itself is a mutable object (like a nested dict or list), both the original and the copy will still point to and share that same inner object. Changing a nested value through one will affect the other. See the Nested Dictionaries section below — a true independent copy of nested structures would require `copy.deepcopy()` instead.

##### 6. Iterating over Dictionaries

```python
# Iterate over keys

for keys in student.keys():
    print(keys)
```

```python
# Iterate over values

for values in student.values():
    print(values)
```

```python
# Iterate over key value pairs

for key,value in student.items():
    print(f"{key}:{value}")
```

##### 7. Nested Dictionaries

```python
#Define

students = {
    "student1" : {"name": "Riad", "age" : 23},
    "student2" : {"name": "Hasan", "age" : 25},
}
```

```python
# Accessing nested Dictionaries elements

print(students["student1"]["name"])  # from student 1
print(students["student2"]["age"])   # from student 2
```

```python
## Iterating over nested dictionaries

for student_id, student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key, value in student_info.items():
        print(f"{key}:{value}")
```

##### 8. Dictionary Comprehension

```python
squares = {x : x**2 for x in range(5)}
print(squares)
```

```python
# Conditional Dictionary Comprehension

#Title : Squares of only even numbers
even = {x : x**2 for x in range(10) if x%2==0}
print(even)
```

##### 9. Practical Examples

```python
# Q1: Use a dictionary to count the frequency of elements in list.

numbers = [1,2,2,3,3,3,4,4,4,4,5,6,10,10] #Given list

frequency = {}
for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number] = 1

print(frequency)
```

```python
# Q2: Merge 2 dictionaries into one

dict1 = {"a":1, "b":2}
dict2 = {"b":3, "c":4}

merged_dict = {**dict1, **dict2}
print(merged_dict)
```

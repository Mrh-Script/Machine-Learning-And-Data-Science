# Tuples
- Tuples are ordered collections of items that are immutable. They are similar to lists, but their immutability makes them different.

```python
## define a tuple

empty_tuple = ()

print(type(empty_tuple))
```

```python
## Another way & list vs tuple

lst = list()
tpl = tuple()

print(type(lst))
print(type(tpl))
```

```python
mixed_tuples = (1, 2, 3, 'a', True, "Dhaka")
print(type(mixed_tuples), mixed_tuples)
```

```python
print(mixed_tuples)

print(mixed_tuples * 3)
```

##### 1. Immutable Nature of Tuples
- Tuples are immutable, means items cannot be changed if once assigned.

```python
## List is Mutable

lst = [1,2,3]
lst[1] = 'a'
print(lst)
```

```python
## Tuples is immutable

tpl = (1,2,3)
tpl[1] = 5
print(tpl)
```

##### 2. Tuples Method
- index(item) : how many occurances 
- count(item) : what is the index of that item

```python
print(tpl)

print(tpl.count(1))
print(tpl.index(3))
```

##### 3. Packing and Unpacking Tuples

```python
## Packing

packed_tuples = (404, "Riad", 'b')
print(packed_tuples)
```

```python
## Unpacking

a, b, c = packed_tuples

print(a)
print(b)
print(c)
```

```python
##  Unpacking with *

numbers = (1,2,3,4,5,6)

first , *middle, last = numbers

print(first)
print(middle)
print(last)
```

##### 4. Nested Tuple

```python
## Nested list

list = [[1,2,3,4], ["Riad", 404], ['a', 'b', 334]]

print(list[0])
print(list[0][-1])

print(list[1])
print(list[1][0])

print(list[-1][:2])
```

```python
## Nested Tuple

nested_tuple = ( (1,2,3,4), ('a', 'b', 'c', 'd'), (True, False))

print(nested_tuple)

## Accessing

print(nested_tuple[0])
print(nested_tuple[1][-1])
```

```python
## Iterating over nested tuples

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item, end= " ")
    print()
```

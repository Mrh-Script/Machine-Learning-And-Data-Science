### 1. For Loop

```python
## for loop in range

for i in range(5):
    print(i)
```

```python
## for loop in string

str = "Uttara University"
for i in str:
    print(i)
```

```python
## Example - 

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

### 2. Nested For Loop

```python
for i in range(2):       # Outer loop
    for j in range(3):   # Inner loop
        print(f"i:{i} and j:{j}")
```

### 3. While Loop

```python
count = 1
while count <= 5:
    print(count)
    count += 1  # Crucial: otherwise go for an infinite loop
```

## 4. Loop Control Statements

#### - Break

```python
# Break : break the loop and stop executing full loop under condition
for i in range(5):
    if i == 3:
        break
    print(i) 
```

#### - Continue

```python
# Continue : Skips items in condition and go for next
for i in range(5):
    if i == 3:
        continue
    print(i) 
```

#### - Pass

```python
## Pass : buffer statement that does noting

for letter in "Python":
    if letter == "h":
        pass  # Placeholder: I'll decide what to do with 'h' later
    print(f"Current letter: {letter}")
```

## Code Examples

**Question :** Print sum upto N natural numbers using While loop.

```python
n = int(input("Enter the value of N : "))

sum = 0
count = 1

while count<=n:
    sum += count
    count = count + 1

print(f"The Sum of first {n} numbers is : ", sum)
```

**Question :** Print sum upto N natural numbers using for loop.

```python
n = int(input("Enter the value of N : "))

sum = 0
for count in range(1, n+1, 1):
    sum += count

print(f"The Sum of first {n} numbers is : ", sum)
```

**Question :** Print the Prime numbers between 1 to N

```python
n = int(input("Enter the value of N : "))

for num in range(1, n+1):

    if num > 1:

        for i in range(2, num):
            if num % i == 0:
                break
        else:
            print(num)
```

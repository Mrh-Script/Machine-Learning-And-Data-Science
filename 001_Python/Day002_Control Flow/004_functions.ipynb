# Python Functions 

## 1. What is a Function, and Why Do We Need It?
A **function** is a named, reusable block of code that performs a specific task. You define it once and can *call* it as many times as you like.

**Why we need functions:**
- **Reusability** — write the logic once, use it many times.
- **Readability** — breaks a big program into small, named pieces.
- **Maintainability** — fix a bug in one place instead of everywhere.
- **Abstraction** — the caller only needs to know *what* the function does, not *how*.

**Basic syntax:**
```python
def function_name(parameters):
    """Docstring"""
    # function body
    return result
```

````python
def greet():
    """A simple function with no parameters."""
    print("Hello! This is my first function.")

# Calling the function
greet()
greet()  # can be reused as many times as needed
````

## 2. Even / Odd Check
First, a plain script (no function) that checks if a number is even or odd, then the same logic wrapped into a **function with one parameter**.

````python
# 2. Even/Odd check WITHOUT a function
number = 7

if number % 2 == 0:
    print(f"{number} is Even")
else:
    print(f"{number} is Odd")
````

### 2.a Even/Odd Check as a Function (One Parameter)

````python
def check_even_odd(number):
    """Takes ONE parameter and prints whether it is even or odd."""
    if number % 2 == 0:
        print(f"{number} is Even")
    else:
        print(f"{number} is Odd")

# Calling the function with different values
check_even_odd(10)
check_even_odd(15)
check_even_odd(0)
````

## 3. Function with Multiple Parameters
A function can accept more than one parameter, separated by commas.

````python
def add_numbers(a, b, c):
    """Takes THREE parameters and returns their sum."""
    total = a + b + c
    print(f"{a} + {b} + {c} = {total}")

add_numbers(5, 10, 15)
add_numbers(1, 2, 3)
````

## 4. Default Parameters
A default parameter has a pre-set value that is used **only if the caller does not provide a value** for it.

````python
def greet_user(name, greeting="Hello"):
    """`greeting` has a default value of 'Hello'."""
    print(f"{greeting}, {name}!")

greet_user("Alice")                # uses default greeting
greet_user("Bob", "Good morning")  # overrides the default
````

## 5. Variable-Length Arguments
Sometimes we don't know in advance how many values will be passed to a function. Python allows this using `*args` (positional) and `**kwargs` (keyword).

### 5.a Positional Variable Arguments — `*args`
`*args` collects any number of **positional** arguments into a tuple.

````python
def add_all(*args):
    """Accepts any number of positional arguments and returns their sum."""
    print("Values received:", args)
    return sum(args)

print(add_all(1, 2, 3))
print(add_all(10, 20, 30, 40, 50))
````

### 5.b Keyword Variable Arguments — `**kwargs`
`**kwargs` collects any number of **keyword** arguments into a dictionary.

````python
def print_profile(**kwargs):
    """Accepts any number of keyword arguments and prints them."""
    print("Profile details received:", kwargs)
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_profile(name="Alice", age=25, city="Dhaka")
````

## 6. Combining `*args` and `**kwargs`
A function can accept normal parameters, `*args`, and `**kwargs` together. The order must be: **normal parameters → \*args → \*\*kwargs**.

````python
def student_info(name, *args, **kwargs):
    """Combines a normal parameter, *args, and **kwargs."""
    print(f"Name: {name}")
    print(f"Extra positional info: {args}")
    print(f"Extra keyword info: {kwargs}")

student_info("Riad", "CSE", "4th Year", roll=332, cgpa=4.00)
````

## 7. The `return` Statement
`return` sends a value (or values) back to the caller and ends the function's execution. Without `return`, a function returns `None` by default.

### 7.a Returning a Single Value

````python
def square(number):
    """Returns a single value."""
    return number ** 2

result = square(6)
print("Square:", result)
````

### 7.b Returning Multiple Values
Python lets a function return multiple values at once — they are packed into a tuple.

````python
def min_max(numbers):
    """Returns MULTIPLE values: the minimum and maximum."""
    return min(numbers), max(numbers)

low, high = min_max([4, 8, 1, 9, 3])
print(f"Minimum: {low}, Maximum: {high}")
````

### 7.c Return Statement with Multiple Parameters

````python
def rectangle_stats(length, width):
    """Takes multiple parameters and returns multiple calculated values."""
    area = length * width
    perimeter = 2 * (length + width)
    return area, perimeter

area, perimeter = rectangle_stats(5, 3)
print(f"Area: {area}, Perimeter: {perimeter}")
````

## Summary
| Concept | Key Idea |
|---|---|
| Function definition | Reusable, named block of code |
| Single parameter | One input value |
| Multiple parameters | Several input values |
| Default parameter | Fallback value if not provided |
| `*args` | Variable number of positional arguments |
| `**kwargs` | Variable number of keyword arguments |
| Combined | Normal params + `*args` + `**kwargs` |
| `return` | Sends value(s) back to the caller |

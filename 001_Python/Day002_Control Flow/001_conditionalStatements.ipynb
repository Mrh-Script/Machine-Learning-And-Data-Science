## The conditional statements are -
- if
- elif
- else
- Nested if

```python
# if - Checks the initial condition

age = 20
if age >= 18:
    print("You are an adult.")
```

```python
# elif - Checks additional conditions if the previous ones were False

temp = 15
if temp > 30:
    print("It's hot.")
else:
    print("It's cool.")
```

```python
# else - Executes if none of the above conditions were True

temp = 15
if temp > 30:
    print("It's hot.")
else:
    print("It's cool.")
```

```python
# Nested if - Checks a condition inside another condition

is_logged_in = True
has_permission = False

if is_logged_in:
    if has_permission:
        print("Welcome to the Admin Dashboard.")
    else:
        print("Login successful, but you lack permissions.")
else:
    print("Please log in first.")
```

# Standard Library Overview 

```python
# array

import array
arr = array.array('i', [1,2,3,4,5])
print(arr)
```

```python
#math

import math
print(math.sqrt(16))
print(math.pi)
```

```python
# random

import random
print(random.randint(1,10))  # random integer
print(random.choice(['apple', 'banana', 'cherry']))  # random choice
```

```python
# File and Directory Access  (os)

import os
print(os.getcwd())    # cwd = current working directory
```

```python
# High Level operations on files and collection of files

import shutil
shutil.copyfile('source.txt', 'destination.txt')
```

```python
# Data Serilization

import json
data = {'name': 'Riad', 'age': 23}  # dictionary
print(type(data))

# Dictionary to String  
json_str = json.dumps(data)
print(json_str)
print(type(json_str))

#String to Dictionary   (Previous / Back into Dictionary)
parsed_data = json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
```

```python
# csv

import csv
with open('example.csv', mode='w',newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['name', 'age'])
    writer.writerow(['Riad', 23])

with open('example.csv', mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
# datetime

from datetime import datetime,timedelta

now = datetime.now()
print(now)

yesterday = now - timedelta(days=1)   # how manyu days previous
print(yesterday)

tomorrow = now + timedelta(days=1)   # how manyu days previous
print(tomorrow)
```

```python
# time

import time

print(time.time())

time.sleep(2)   # 2 seconds
print(time.time())
```

```python
# regular expression

import re
pattern = r'\d+'   # d = digit
text = 'There are 123 apples'
match = re.search(pattern, text)
print(match.group())
```
